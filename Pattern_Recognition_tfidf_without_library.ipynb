## Tujuan

Tujuan dari notebook ini adalah mengimplementasikan proses preprocessing teks
dan ekstraksi fitur TF-IDF secara manual tanpa menggunakan library NLP.
Metode ini digunakan untuk merepresentasikan dokumen teks dalam bentuk
vektor numerik sehingga dapat digunakan dalam berbagai aplikasi text mining.

## 1. Import Library Dasar Python

```python
import math
```

## 2. Membuat Dataset

```python
documents = [
    "Machine learning adalah cabang dari kecerdasan buatan yang memungkinkan komputer untuk belajar dari data tanpa harus diprogram secara eksplisit. Dengan menggunakan algoritma tertentu, komputer dapat menemukan pola dalam data dan membuat prediksi atau keputusan secara otomatis.",
    "Data mining merupakan proses untuk menemukan pola atau informasi yang berguna dari kumpulan data yang sangat besar. Teknik ini sering digunakan dalam berbagai bidang seperti bisnis, kesehatan, dan penelitian ilmiah untuk membantu pengambilan keputusan berdasarkan data.",
    "Dalam bidang ilmu komputer, machine learning dan data mining sering digunakan bersama untuk menganalisis data yang kompleks. Kedua teknik ini membantu peneliti untuk memahami pola yang tersembunyi dalam data dan menghasilkan pengetahuan baru."
]

print("Jumlah dokumen:", len(documents))
```

```python
print("\n" + "="*60)
print("                    DATASET DOKUMEN")
print("="*60)

for i, doc in enumerate(documents):

    print(f"\nDokumen {i+1}")
    print("-"*40)
    print(doc)
```

## 3. Membuat Stopwords Manual

```python
stopwords = [

"yang",
"dan",
"di",
"ke",
"dari",
"untuk",
"dengan",
"dalam",
"atau",
"pada",
"adalah",
"ini",
"itu",
"sebagai",
"secara",
"dapat",
"akan",
"agar",
"oleh",
"karena",
"saat",
"sering",
"lebih",
"juga",
"tanpa",
"telah",
"para",
"suatu",
"antara"

]
```

## 4. Membuat Fungsi Preprocessing




1. Case Folding
2. Cleaning
3. Tokenization
4. Stopword Removal





```python
def preprocess(text):

    # case folding
    text = text.lower()

    # cleaning karakter
    characters = ",.!?;:()[]{}0123456789"

    for c in characters:
        text = text.replace(c, "")

    # tokenization
    tokens = text.split()

    # stopword removal
    filtered = []

    for word in tokens:

        if word not in stopwords:
            filtered.append(word)

    return filtered
```

## 5. Melakukan Preprocessing Semua Dokumen


```python
processed_docs = []

for doc in documents:

    processed = preprocess(doc)

    processed_docs.append(processed)

print("\n" + "="*60)
print("                 HASIL PREPROCESSING")
print("="*60)

for i, doc in enumerate(processed_docs):

    print(f"\nDokumen {i+1}")
    print("-"*30)

    for word in doc:
        print(word, end=" | ")

    print()
```

## 6. Membuat Vocabulary
Semua kata unik dari corpus.

```python
vocab = []

for doc in processed_docs:

    for word in doc:

        if word not in vocab:
            vocab.append(word)

print("\n" + "="*60)
print("                     VOCABULARY")
print("="*60)

print("Jumlah kata unik:", len(vocab))
print("-"*60)

for i, word in enumerate(vocab, 1):

    print(f"{i:>2}. {word}")
```

## 7. Menghitung Term Frequency (TF)
Jumlah kemunculan kata dalam dokumen.

```python
def compute_tf(doc, vocab):

    tf_dict = {}

    for word in vocab:

        tf_dict[word] = doc.count(word)

    return tf_dict
```

```python
tf_list = []

for doc in processed_docs:

    tf = compute_tf(doc, vocab)

    tf_list.append(tf)

print("\n" + "="*60)
print("                   TERM FREQUENCY")
print("="*60)

for i, tf in enumerate(tf_list):

    print(f"\nDokumen {i+1}")
    print("-"*40)

    for word, count in tf.items():

        if count > 0:
            print(f"{word:<15} : {count}")
```

## 8. Menghitung IDF
Rumus: IDF = 1 + log(N / df)  
N = jumlah dokumen  
df = jumlah dokumen yang mengandung term

```python
def compute_idf(processed_docs, vocab):

    N = len(processed_docs)

    idf_dict = {}

    for word in vocab:

        df = 0

        for doc in processed_docs:

            if word in doc:
                df += 1

        idf = 1 + math.log(N / df)

        idf_dict[word] = idf

    return idf_dict
```

```python
idf = compute_idf(processed_docs, vocab)

print("\n" + "="*50)
print("              IDF VALUE")
print("="*50)

for word, value in idf.items():
    print(f"{word:<15} : {value:.3f}")
```

## 9. Menghitung TF-IDF
TF-IDF (Term Frequency - Inverse Document Frequency) digunakan untuk
mengukur seberapa penting suatu kata dalam dokumen terhadap keseluruhan
dokumen dalam corpus.

TF menunjukkan frekuensi kemunculan kata dalam dokumen, sedangkan
IDF mengukur seberapa jarang kata tersebut muncul pada seluruh dokumen.

TF-IDF = TF × IDF

```python
def compute_tfidf(tf, idf):

    tfidf = {}

    for word in tf:

        tfidf[word] = tf[word] * idf[word]

    return tfidf
```

```python
tfidf_list = []

for tf in tf_list:

    tfidf = compute_tfidf(tf, idf)

    tfidf_list.append(tfidf)
```

## 10. Menampilkan TF-IDF per Dokumen

```python
print("\n" + "="*50)
print("            TF-IDF PER DOKUMEN")
print("="*50)

for i in range(len(tfidf_list)):

    print(f"\nDokumen {i+1}")
    print("-"*30)

    for word, value in tfidf_list[i].items():

        if value > 0:
            print(f"{word:<15} : {value:.3f}")
```
